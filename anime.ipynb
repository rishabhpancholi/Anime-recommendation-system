```python
import numpy as np
import pandas as pd
```

```python
anime=pd.read_csv('anime.csv')
rating=pd.read_csv('rating.csv')
anime.tail()
```

```python
rating_with_name=rating.merge(anime,on='anime_id')
```

```python
rating_with_name.head()
```

```python
rating_with_name = rating_with_name.drop('rating_y', axis=1)
```

```python
num_ratings_df=rating_with_name.groupby('name').count()['rating_x'].reset_index()
num_ratings_df=num_ratings_df.rename(columns={'rating_x':'num_ratings'})
num_ratings_df
```

```python
avg_ratings_df=rating_with_name.groupby('name').mean()['rating_x'].reset_index()
avg_ratings_df=avg_ratings_df.rename(columns={'rating_x':'avg_ratings'})
avg_ratings_df
```

```python
popular_df=avg_ratings_df.merge(num_ratings_df,on='name')
```

```python
popular_df=popular_df[popular_df['num_ratings']>=250].sort_values('avg_ratings',ascending=False).head(50)
```

```python
popular_df=popular_df.merge(anime,on='name').drop_duplicates('name')[['name','genre','type','num_ratings','avg_ratings','anime_id']]
popular_df
```

```python
anime=anime[['anime_id','name','genre','type']]
```

```python
anime.isnull().sum()
```

```python
anime.dropna(inplace=True)
```

```python
anime.drop_duplicates(inplace=True)
```

```python
anime['genre']=anime['genre'].apply(lambda x:x.split(','))
anime['type']=anime['type'].apply(lambda x:x.split(' '))
```

```python
anime['tags']=anime['genre']+anime['type']
```

```python
anime=anime[['anime_id','name','tags']]
```

```python
anime['tags']=anime['tags'].apply(lambda x:[i.lower() for i in x])
```

```python
def list_to_string(my_list):
  my_str=''
  for i in my_list:
      if my_str!='':
          my_str+=','
      my_str+=i

  return my_str
```

```python
anime['tags']=anime['tags'].apply(lambda x:list_to_string(x))
```

```python
import nltk
```

```python
from nltk.stem import SnowballStemmer
import pandas as pd

stemmer = SnowballStemmer('english')

def stem(text):
    return " ".join([stemmer.stem(word) for word in text.split()])
    
anime['tags'] = anime['tags'].fillna('').apply(stem)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=10000,stop_words='english')
```

```python
tfidf_matrix = tfidf.fit_transform(anime['tags'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
def recommend(title, cosine_sim=cosine_sim):
    idx = anime[anime['name'] == title].index[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:6]  
    movie_indices = [i[0] for i in sim_scores]
    return anime['name'].iloc[movie_indices]
```
